## 5. 🧠 Choose Your YOLO11 Model
Ultralytics offers several sizes of the YOLO11 model. You can switch between them by changing the model name in the training command.

- `yolo11n.pt` (Nano) - Fastest, smallest, great for edge devices (mobile/Raspberry Pi).
- `yolo11s.pt` (Small) - Good balance of speed and accuracy.
- `yolo11m.pt` (Medium) - Better accuracy, slightly slower.
- `yolo11l.pt` (Large) - High accuracy, requires more GPU memory.
- `yolo11x.pt` (Xtra Large) - Highest accuracy, slowest, requires powerful GPUs.

## 6. 🚀 Train the YOLO11 Model
Now we train the model! We will save the project directly to Google Drive so that if Colab disconnects, you don't lose your progress.

**⚠️ Note:** Update `data='/content/drive/MyDrive/dataset/data.yaml'` to point to YOUR `data.yaml` file.

````python
from ultralytics import YOLO

# 1. Load a pre-trained model (we recommend nano 'yolo11n.pt' or small 'yolo11s.pt' to start)
model = YOLO('yolo11n.pt')

# 2. Train the model
# - data: path to your data.yaml
# - epochs: how many times to iterate over the dataset (start with 50-100)
# - imgsz: image size (640 is standard)
# - batch: batch size (use 16 or 32 for T4 GPU. Lower this if you get Out Of Memory errors)
# - project: where to save the training results
results = model.train(
    data='/content/drive/MyDrive/dataset/data.yaml', 
    epochs=50, 
    imgsz=640, 
    batch=16,
    project='/content/drive/MyDrive/YOLO11_Training',
    name='my_custom_model',
    device=0  # Forces training to use the GPU
)
````

## 7. 📊 Evaluate the Model
Once training is finished, we can evaluate its performance (mAP, Precision, Recall) on the validation set.

````python
# Load the best weights from your training session
model = YOLO('/content/drive/MyDrive/YOLO11_Training/my_custom_model/weights/best.pt')

# Run validation
metrics = model.val()
print(f"Mean Average Precision (mAP50-95): {metrics.box.map}")
````

## 8. 📸 Run Inference (Test your model!)
Let's test the model on a new image or video to see how well it learned.

````python
# Provide the path to a test image or video
test_source = '/content/drive/MyDrive/dataset/images/val/test_image.jpg' # UPDATE THIS PATH

# Run prediction
results = model.predict(source=test_source, save=True, conf=0.5)

# Display the predicted image
import cv2
from google.colab.patches import cv2_imshow
import os

# Ultralytics saves outputs to runs/detect/predict by default
# We will display the first result
if os.path.exists(results[0].save_dir):
    img_path = os.path.join(results[0].save_dir, os.path.basename(test_source))
    img = cv2.imread(img_path)
    if img is not None:
        cv2_imshow(img)
````

## 10. 🛠️ Troubleshooting Common Errors

| Error | Solution |
|-------|----------|
| **CUDA not available** | You are on a CPU runtime. Go to `Runtime > Change runtime type` and select T4 GPU. |
| **Dataset not found** | Double-check the paths in your `data.yaml`. Ensure they use absolute paths (e.g., `/content/drive/MyDrive/...`) rather than relative paths. |
| **Out of memory (OOM)** | The GPU ran out of VRAM. Lower the `batch` size in `model.train()` from 16 to 8 or 4. Alternatively, use a smaller model like `yolo11n.pt`. |
| **FileNotFoundError: data.yaml** | Google Drive might not be mounted correctly, or the path is misspelled. Run `!ls /content/drive/MyDrive` to explore your drive. |
